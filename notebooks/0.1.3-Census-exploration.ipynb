```python
API_KEY = "ebc7fa75b662b431686dfa3e6917b77efaad8853"
YEAR = "2022" 
# Using ACS 5-Year Data Profiles for pre-calculated rates/percentages
BASE_URL = f"https://api.census.gov/data/{YEAR}/acs/acs5/profile"

# Mapping of requested concepts to Data Profile (DP) variables
# Note: E = Estimate, PE = Percentage Estimate
VARIABLES = {
    "NAME": "Geography Name",
    "DP05_0001E": "Total population",
    "DP05_0005PE": "Population 5 to 19 years (Proxy for 5-17)",
    "DP05_0009PE": "Population 20 to 64 years (Proxy for 18-64)",
    "DP05_0024PE": "Population 65+",
    "DP02_0072E": "Veterans population",
    "DP03_0119PE": "Poverty rate (all ages)",
    "DP03_0128PE": "Child poverty rate (<18 years)",
    "DP03_0062E": "Median household income",
    "DP03_0074PE": "SNAP participation rate",
    "DP04_0142PE": "Housing cost burden (>35% income, proxy for 30%)",
    "DP02_0068PE": "Educational attainment (bachelor's+)",
    "DP03_0099PE": "Uninsured rate",
    "DP02_0067PE": "High school graduation rate",
    "DP03_0051PE": "Government employment",
    "DP02_0093PE": "Foreign-born population percentage",
    "DP02_0114PE": "Language spoken at home (non-English)",
    "DP02_0044E": "Grandparents responsible for grandchildren",
    "DP04_0046PE": "Homeownership rate",
    "DP04_0003PE": "Vacant housing units",
    "DP02_0153PE": "Broadband internet access",
    "DP02_0152PE": "Computer at home"
}

params = {
    "get": ",".join(VARIABLES.keys()),
    "for": "county:*", # Adjust geography (e.g., state:*)
    "key": API_KEY
}

response = requests.get(BASE_URL, params=params)

if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data[1:], columns=data[0])
    df.rename(columns=VARIABLES, inplace=True)
    print(df.head())
else:
    print(f"Error: {response.text}")
```
